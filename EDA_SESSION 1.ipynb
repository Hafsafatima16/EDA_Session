**CREATE DATA FRAME USING LIST**

```python
names=["ramesh","suresh","satish"]
age=[30,40,50]
names,age
```

```python
pd.DataFrame()       #to make the dataframe
```

*PROVIDE DATA*

```python
pd.DataFrame(zip(names,age))
```

```python
#providing the columns 
pd.DataFrame(zip(names,age),columns=["name","age"])
```

```python
#providing index
pd.DataFrame(zip(names,age),columns=["name","age"],index=['a','b','c'])
```

```python
df=pd.DataFrame(zip(names,age),columns=["name","age"],index=['a','b','c'])
df
```

```python
city_names=['hyd','banglore','mumbai']
df['city']=city_names
```

```python
df
```

**DROP THE COLUMN OR UPDATE THE EXISTING ONE**

- In order to drop the column
- We need to use drop method
- It takes 3 parameters
  - drop column or row
  - mention the column name
  - axis
    - axis=1 reference as column
    - axis=0 reference as row
- you want to create a new dataframe or
- you want overwrite the existing dataframe
  - inplace= True

```python
df.drop('city',    #column name
       axis=1,     #column
       inplace=True)  #update the existing data
df
```

```python
df.drop('c',     #row name
       axis=0,   #row
       inplace=True)
df
```
