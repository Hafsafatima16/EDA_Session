$OUTLIER-ANALYSIS$

- Step-1: Find the Q1 , Q2 and Q3
    - np.percentile(column data, q)
- Step-2: Calculate the IQR
    - IQR= Q3-Q1
- Step-3: Calculate lower boundary and upper boundary
    - lb: Q1-1.5IQR
    - ub: Q3+1.5IQR
- Step-4: Find the Outliersdf
    - c1: column data < lb
    - c2: column data > ub
    - c: apply the main condition
    - main data[c]

**Step - 1**

```python
# FINDING Q1, Q2, Q3 

Q1 = np.percentile(lpred_df['ApplicantIncome'],25)
Q2 = np.percentile(lpred_df['ApplicantIncome'],50)
Q3 = np.percentile(lpred_df['ApplicantIncome'],75)
```

**Step - 2**

```python
# CALCULATE IQR

IQR = Q3 - Q1
```

**Step - 3**

```python
# CALCULATE LOWER AND UPPER BOUNDRIES

lower_b = Q1 - 1.5 * IQR
upper_b = Q3 + 1.5 * IQR
```

**Step - 4**

```python
c1 = lpred_df['ApplicantIncome']<lower_b
c2 = lpred_df['ApplicantIncome']>upper_b
c = c1 | c2
outlier_df = lpred_df[c]
outlier_df
```

**DEALING THE OUTLIERS**

**Drop the outliers**

- we can drop the outliers if outlier percentage < 2%
- But this is not recommended, we might lose other columns data also.

**Impute with Median values**

- As we know that Median is not affected by outliers
- So it is good prcatice to impute outliers with Median value

**Cap with Q3 or Q1 value**

- If outliers are presnt less than lower bound then fill with Q1
- If outliers are more than upper bound then fill with Q3

*Impute with median values*

```python
#-----------------step-1-------------------------------#
Q1 = np.percentile(lpred_df['ApplicantIncome'],25)
Q2 = np.percentile(lpred_df['ApplicantIncome'],50)
Q3 = np.percentile(lpred_df['ApplicantIncome'],75)
#-----------------step-2-------------------------------#
IQR = Q3 - Q1
#-----------------step-3--------------------------------#
lower_b = Q1 - 1.5 * IQR
upper_b = Q3 + 1.5 * IQR
#-----------------step-4--------------------------------#
list1 = []
median = lpred_df['ApplicantIncome'].median()
for i in lpred_df['ApplicantIncome']:
    if i < lower_b or i > upper_b:
        list1.append(median)
    else:
        list1.append(i)
        
lpred_df['ApplicantIncome_new']=list1
```

**np.where**

- np.where(con,True,False)
- Will take 3 arguments
    - Condition
    - True value
    - False value

```python
Q1 = np.percentile(lpred_df['ApplicantIncome'],25)
Q2 = np.percentile(lpred_df['ApplicantIncome'],50)
Q3 = np.percentile(lpred_df['ApplicantIncome'],75)

IQR = Q3 - Q1

lower_b = Q1 - 1.5 * IQR
upper_b = Q3 + 1.5 * IQR

list1 = []
median = lpred_df['ApplicantIncome'].median()

c1 = lpred_df['ApplicantIncome'] < lower_b
c2 = lpred_df['ApplicantIncome'] > upper_b
con = c1 | c2

true = median
false = lpred_df['ApplicantIncome']

lpred_df['ApplicantIncome'] = np.where(con, true, false)
plt.boxplot(lpred_df['ApplicantIncome'])
plt.show()
```
