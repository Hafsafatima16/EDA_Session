$TAKE - loc - iloc$

```python
df.take([2,3,4])    #GIVES SELECTED NUMBER OF ROWS
```

- Signature: df.take(indices, axis: 'Axis' = 0, **kwargs) -> 'NDFrameT'
- Docstring:Return the elements in the given *positional* indices along an axis.

```python
df.take([2,3,4],axis=1)     #GIVES SELECTED NUMBER OF COLUMNS
```

```python
df.take((100,200,300))
```

$iloc$

- df.iloc[<rows>,<columns>]<br>
- df.iloc[<rows_start:rows_end>,<columns_start:column_end>]<br>
- row=[]<br>
- column=[]<br>
- df.iloc[rows,columns]

```python
df.iloc[5:10]
```

```python
df.iloc[5:10,2:5]
```

```python
df.iloc[:,2:5]
```

```python
df.iloc[5:10,:]  #specific row and all the columns
```

```python
df.iloc[100:400,4:7]
```

```python
rows=[100,200,300]
column=[4,7,8]
df.iloc[rows,column]
```

```python
df.iloc[[100,200,300],[4,7,8]]
```
