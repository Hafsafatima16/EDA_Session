$DATA-TRANSFORMATION-TECHNIQUES$

- Generally used for to convert Normal distribution
- Because all statistical math analysis by assumption Data follows Normal distribution
- It also avoid skew ness.
- We have some important transformation:
    - Log transformation
    - Exponential transformation
    - Reciprocal transformation
    - Square root transformation
    - Power transformaton

**Step - 1**

**Read the required packages**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Step - 2**

**Read the data**

$Exponential-Data$

```python
exp_data = np.random.exponential(size=10000)
exp_data[:10]
```

```python
# We are consider exponential data set with random 1k observations
# using numpy package
```

```python
plt.hist(exp_data,bins=30,label='Exponential')
plt.legend()
plt.show()
```

$Norm-Data$

```python
norm_data = np.random.normal(size=10000)
norm_data[:10]
```

```python
plt.hist(norm_data,bins=30,label='Normal')
plt.legend()
plt.show()
```

**STEP - 3**

$Log-TRANSFORMATION$

- np.log is used for log transformation
- Generally log transformation will not convert data into normal
- It avoids skew ness
- np.log means natural logorithm base=e

```python
x=2
np.log(2)
```

```python
log_trns = np.log(exp_data)
log_trns[:10]
```

```python
exp_data[:10]
```

```python
plt.hist(log_trns,bins=30,label='Log Transform')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,
                                bins=30,
                                label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_trns,
                       bins=30,
                       label='Log Transform')
plt.legend()
plt.show()
```

**Step - 4**

$Reciprocal-Transformation$

> Reciprocal transformation fails when data has zero value

```python
exp_data[:10]
```

```python
rec_data = np.reciprocal(exp_data)
rec_data[:10]
```

```python
plt.hist(rec_data,bins=30,label='Reciprocal')
plt.legend()
plt.show()
```

**Step - 5**

$Square-Root-Transformation$

```python
print(25**2) # square,
print(25**(1/2)) # square root
print(np.sqrt(25))
```

```python
sqrt_data = np.sqrt(exp_data)
plt.hist(sqrt_data,
        bins=30,
        label='Square root')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,
                                bins=30,
                                label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(sqrt_data,
                       bins=30,
                       label='Square root')
plt.legend()
plt.show()
```

```python
sns.displot(sqrt_data)
```

**Step - 6**

$Power-Transformation$

- It is related to sklearn package
- Package name: **sklearn.preprocessing**
- Method name: **Power Transformer**
- Inside: **Box-Cox , yeo-jhonson**

```python
exp_data
```

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer()
power_data = pt.fit([exp_data]).transform([exp_data])
power_data
```

```python
plt.hist(power_data[0])
plt.show()
```
